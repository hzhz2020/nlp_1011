### Hyper parameter tunning:

### 1. Varying the size of the hidden dimension of the CNN and RNN


### CNN

```python
CNN150 = pkl.load(open('./parameters_tunning/SNL_CNN_HIDDEN150.p','rb'))
CNN300 = pkl.load(open('./parameters_tunning/SNL_CNN_HIDDEN300.p','rb'))
CNN450 = pkl.load(open('./parameters_tunning/SNL_CNN_HIDDEN450.p','rb'))
```

```python
# Plot learning curve  --- ALL TRAINED SAME NUMBER OF EPOCH
CNN150_TRAIN_ACC = CNN150['train_accuracy']
CNN150_VAL_ACC = CNN150['val_accuracy']

CNN300_TRAIN_ACC = CNN300['train_accuracy']
CNN300_VAL_ACC = CNN300['val_accuracy']

CNN450_TRAIN_ACC = CNN450['train_accuracy']
CNN450_VAL_ACC = CNN450['val_accuracy']
```

### plot training and validation accuracy

```python
fig = plt.figure(figsize = (12,8))

total_point = len(CNN450_TRAIN_ACC)
#training accuracy
plt.plot(list(range(1,total_point+1)), CNN450_TRAIN_ACC, label = 'Train Accuracy, CNN_HiddenSize450')
plt.plot(list(range(1,total_point+1)), CNN300_TRAIN_ACC, label = 'Train Accuracy, CNN_HiddenSize300')
plt.plot(list(range(1,total_point+1)), CNN150_TRAIN_ACC, label = 'Train Accuracy, CNN_HiddenSize300')

plt.plot(list(range(1,total_point+1)), CNN450_VAL_ACC, label = 'Val Accuracy, CNN_HiddenSize450')
plt.plot(list(range(1,total_point+1)), CNN300_VAL_ACC, label = 'Val Accuracy, CNN_HiddenSize300')
plt.plot(list(range(1,total_point+1)), CNN150_VAL_ACC, label = 'Val Accuracy, CNN_HiddenSize150')

plt.xlabel('300 iterations')
plt.ylabel('Accuracy')

plt.title('Varying hidden dimension size of CNN')
plt.legend(loc='best')
plt.show
```

```python
print('CNN hidden size 450, best validation accuracy: {}'.format(max(CNN450_VAL_ACC)))
print('CNN hidden size 300, best validation accuracy: {}'.format(max(CNN300_VAL_ACC)))
print('CNN hidden size 150, best validation accuracy: {}'.format(max(CNN150_VAL_ACC)))
```

```python
print('CNN hidden size 450, trainable parameters: {}'.format(CNN450['trainable_parameters']))
print('CNN hidden size 300, trainable parameters: {}'.format(CNN300['trainable_parameters']))
print('CNN hidden size 150, trainable parameters: {}'.format(CNN150['trainable_parameters']))
```

### RNN

```python
RNN150 = pkl.load(open('./parameters_tunning/SNL_RNN_HIDDEN150.p','rb'))
RNN300 = pkl.load(open('./parameters_tunning/SNL_RNN_HIDDEN300.p','rb'))
RNN450 = pkl.load(open('./parameters_tunning/SNL_RNN_HIDDEN450.p','rb'))
```

```python
RNN150_TRAIN_ACC = RNN150['train_accuracy']
RNN150_VAL_ACC = RNN150['val_accuracy']

RNN300_TRAIN_ACC = RNN300['train_accuracy']
RNN300_VAL_ACC = RNN300['val_accuracy']

RNN450_TRAIN_ACC = RNN450['train_accuracy']
RNN450_VAL_ACC = RNN450['val_accuracy']
```
